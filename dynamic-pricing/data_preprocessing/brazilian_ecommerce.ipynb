```python
import pandas as pd
import collections
from tqdm import tqdm
```

```python
df_order_item = pd.read_csv('/Users/ybeen/yb/대회/KURLY_HACK_FESTA/Data/brazilian_e-commerce/olist_order_items_dataset.csv')
df_order_customer = pd.read_csv('/Users/ybeen/yb/대회/KURLY_HACK_FESTA/Data/brazilian_e-commerce/olist_orders_dataset.csv')
```

```python
df_order_item.head()
```

```python
df_order_customer.head()
```

```python
df = pd.merge(left = df_order_item , right = df_order_customer, how = "left", on = "order_id")
```

```python
df_use = df[['order_id', 'product_id', 'customer_id', 'price']]
```

```python
df_use
```

```python
repurchase = {}
for _, c_id in enumerate(tqdm(df_use['customer_id'].unique())):
    repurchase[c_id] = dict(collections.Counter(df_use[df_use['customer_id']==c_id]['product_id']))
```

```python
repurchase
```

```python
df_repurchase = pd.DataFrame(columns=['customer_id', 'product_id', 'repurchase'], data=repurchase)
```

```python
repurchase
```

```python
r_list = []

for c_id in repurchase.keys():
    for p_id in repurchase[c_id].keys():
        r_list.append([c_id, p_id, repurchase[c_id][p_id]])
```

```python
df_purchase = pd.DataFrame(columns=['customer_id', 'product_id', 'purchase'], data=r_list)
```

```python
df_purchase.to_csv('customer_product_purchase.csv', index=False)
```

```python
df_count = pd.merge(left = df_count , right = df_repurchase[['product_id', 'purchase']], how = "left", on = "product_id")
df_count = df_count.fillna(0)
df_count['repurchase'] = df_count['purchase_y']/df_count['purchase_x']
```

```python
df_count = df_count.fillna(0)
df_count[df_count['repurchase']!=0]
```

```python
df_count = df_count.drop(['purchase'], axis=1)
df_count.rename(columns = {'purchase_x' : 'purchase_total', 'purchase_y' : 'purchase_twice'}, inplace = True)
df_count = df_count.astype({'purchase_twice':int})
```
