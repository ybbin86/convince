```python
import pandas as pd
```

```python
df = pd.read_csv('/Users/ybeen/Downloads/Bas_GarakPrice_배.csv', encoding='cp949')
```

```python
df.head()
```

```python
df['품목명'] = df['품목명'].apply(lambda x: x[10:])
df['품목명'] = df['품목명']+'_'+df['단량'].astype(str)
df = df.drop(['단량'], axis='columns')
```

```python
with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    print(df[(df['등급']=='특')].sort_values(by=['일자']))
```

```python
df_mean = pd.DataFrame()
for y in range(200300,202200, 100):
    for m in range(1,13):
        date = str(y+m)
        df1 = df[(df['등급']=='특') & (df['일자'].str.contains(date))].groupby('품목명').mean()
        df2 = df[(df['등급']=='상') & (df['일자'].str.contains(date))].groupby('품목명').mean()
        df3 = df[(df['등급']=='중') & (df['일자'].str.contains(date))].groupby('품목명').mean()
        df4 = df[(df['등급']=='하') & (df['일자'].str.contains(date))].groupby('품목명').mean()

        df1['등급'] = '특'
        df2['등급'] = '상'
        df3['등급'] = '중'
        df4['등급'] = '하'

        temp = pd.concat([df1,df2,df3,df4])
        temp['일자'] = date
        df_mean = pd.concat([df_mean,temp])
df_mean = df_mean.reset_index()
```

```python
df_mean.to_csv('pear.csv', index=False)
```
